```python
import numpy as np
import keras# an easier way to do perceptron and Neoral network
import matplotlib.pyplot as plt
from keras.models import Sequential# from subclass of models import sequential so we can later define the layers of Neoral network
from keras.layers import Dense#in dense class ever note in th dense layer is connected to the preceding layer
from keras.optimizers import Adam# importing adam from sub package of optimizer
#Adam optimization algorithm is a combination of two stichastic optimizing method:Adagrad+RMSprop
#we use stochastic gradint descent which is a lot faster regular gradient descent as it was calculationg gradient on all the points
#as we want to reachthe min we call it optimizer, minimizing the error 
%matplotlib inline
```

```python
n_pts = 500
np.random.seed(0)
Xa = np.array([np.random.normal(13, 2, n_pts),
               np.random.normal(12, 2, n_pts)]).T
Xb = np.array([np.random.normal(8, 2, n_pts),
               np.random.normal(6, 2, n_pts)]).T
 
X = np.vstack((Xa, Xb))
y = np.matrix(np.append(np.zeros(n_pts), np.ones(n_pts))).T
 
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
```

```python
model = Sequential()# we specify a variable
model.add(Dense(units=1, input_shape=(2,), activation='sigmoid')) # we add layers to our neoral ntworks
# in above line we add properties of dense layer, units =1 as we have 1 perceptron, we hav 2 input nodes, activation belongs to output layer which we wanted to be the probability of segmoid
#we have specified the layrs of input and output. to complie a keras model we need an optimizer
adam=Adam(lr = 0.1 )#this is the learning rate(should not be to small or too large)
model.compile(adam, loss='binary_crossentropy', metrics=['accuracy'])
h=model.fit(x=X, y=y, verbose=1, batch_size=50,epochs=500, shuffle='true')#epoch=period(every time entire our data points), we defined our training data labled
#verbose=0 yani hich resultiyo print nemikone bebinim che khabare,1 epoch is too big to feed it to our computer all at once so we use smaller batch sizes, 500 top, 500 down batch size 50 so it will take 50 iterations for
#-complete n epoch, epochs=number of steps and periods 500 times doing an epoch (entire dataset) if epochs be too large then we have overfitting
#shuffle = true it will shuffle our dataset each time after an epoch, it can help to not to get stuck in local minimums
#the fit function is for returning a history object, recording of the training
```

```python
plt.plot(h.history['accuracy'])
 
plt.legend(['accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
 
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.title('loss')
plt.xlabel('epoch')
```

```python
def plot_decision_boundary(X, y, model):#model is what we made earlier containing all neoral network eqs, plot the line=decision boundary that seperates our data
    #X (having 2 dimensions) is the array of data, y is the lable of data, and our neoral network model 
    x_span = np.linspace(min(X[:,0]) - 1, max(X[:,0]) + 1)#we need to define a spine of x and y coordinates in our dataset
    #is equal to equally spaced points between th absulote min and max of horizental value in range of X[:,0], print(x_span) will give us 50 points between min and max
    #-1 and +1 are just some tolerence to have more spacious grid for our data
    y_span = np.linspace(min(X[:,1]) - 1, max(X[:,1]) + 1)
    xx, yy = np.meshgrid(x_span, y_span)# xx and yy 2 dikmentional arrays. what's meshgrid? it returns 2D x_span 50 by 50 for xx, similarly for yy
    #meshgrid makes a 2D array out of a 1D array
    xx_, yy_ = xx.ravel(), yy.ravel()
    grid = np.c_[xx_, yy_]#simply we made a grid for the model to predict the best seperation line
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
 
plot_decision_boundary(X, y, model)
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
x = 7.5
y = 5
 
 
point = np.array([[x, y]])
prediction = model.predict(point)# here we want to see if the model can predict the probability of a new cordinate point to be lable 1 
plt.plot([x], [y], marker='o', markersize=10, color="red")# it will show the point on the graph
print("prediction is: ",prediction)
```
