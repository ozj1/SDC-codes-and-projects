```python
  #very important: from here everything is like deep neural network, so conventional neural network will come before deep learning neural network which uses dense layers

import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Dense
from keras.layers import Flatten, Dropout
from keras.utils.np_utils import to_categorical
from keras.layers.convolutional import Conv2D, MaxPooling2D
import random
import pickle
import pandas as pd #to import data with csv format
import cv2
 
 
from keras.callbacks import LearningRateScheduler, ModelCheckpoint
 
%matplotlib inline
```

```python
np.random.seed(0)
```

```python
# TODO: Implement load the data here.
with open('german-traffic-signs/train.p', 'rb') as f:
    train_data = pickle.load(f)
with open('german-traffic-signs/valid.p', 'rb') as f:
    val_data = pickle.load(f)
# TODO: Load test data
```

```python
with open('german-traffic-signs/test.p', 'rb') as f:
    test_data = pickle.load(f)
 
    
# Split out features and labels
X_train, y_train = train_data['features'], train_data['labels']
X_val, y_val = val_data['features'], val_data['labels']
X_test, y_test = test_data['features'], test_data['labels']
 
#already 4 dimensional
print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
```

```python
# STOP: Do not change the tests below. Your implementation should pass these tests. 
assert(X_train.shape[0] == y_train.shape[0]), "The number of images is not equal to the number of labels."
assert(X_train.shape[1:] == (32,32,3)), "The dimensions of the images are not 32 x 32 x 3."
assert(X_val.shape[0] == y_val.shape[0]), "The number of images is not equal to the number of labels."
assert(X_val.shape[1:] == (32,32,3)), "The dimensions of the images are not 32 x 32 x 3."
assert(X_test.shape[0] == y_test.shape[0]), "The number of images is not equal to the number of labels."
assert(X_test.shape[1:] == (32,32,3)), "The dimensions of the images are not 32 x 32 x 3."
```

```python
data = pd.read_csv('german-traffic-signs/signnames.csv')
  
num_of_samples=[]
 
cols = 5
num_classes = 43
 
fig, axs = plt.subplots(nrows=num_classes, ncols=cols, figsize=(5,50))
fig.tight_layout()
 
for i in range(cols):
    for j, row in data.iterrows():
      x_selected = X_train[y_train == j]
      axs[j][i].imshow(x_selected[random.randint(0,(len(x_selected) - 1)), :, :], cmap=plt.get_cmap('gray'))
      axs[j][i].axis("off")
      if i == 2:
        axs[j][i].set_title(str(j) + " - " + row["SignName"])
        num_of_samples.append(len(x_selected))
```

```python
import cv2#as a tool to manipulate over images
 
plt.imshow(X_train[1000])
plt.axis("off")
print(X_train[1000].shape)
print(y_train[1000])
```

```python
def grayscale(img):#making all images in grayscale, first data preprocessing section
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img
```

```python
img = grayscale(X_train[1000])
plt.imshow(img)
plt.axis("off")
print(img.shape)
```

```python
def equalize(img):# the second data preprocessing section to normalize the images pixel density histograms
#this func will increase the contrast in the picture making it easier for the code to learn (by normalizing the histogram, pixels intesities will be distributed all over the picture causing an increase in image contrast)
    img = cv2.equalizeHist(img)#note that this func only accepts images without any depth meaning 2D
    return img
```

```python
img = equalize(img)
plt.imshow(img)
plt.axis("off")
print(img.shape)
```

```python
def preprocess(img):
    img = grayscale(img)
    img = equalize(img)
    img = img/255
    return img
```

```python
#updating the entire dataset
X_train = np.array(list(map(preprocess, X_train)))
X_test = np.array(list(map(preprocess, X_test)))
X_val = np.array(list(map(preprocess, X_val)))
 
plt.imshow(X_train[random.randint(0, len(X_train) - 1)])
plt.axis('off')
print(X_train.shape)
```

```python
#in order to us convlotional neural network we need to add depth to the images
X_train = X_train.reshape(34799, 32, 32, 1)#34799 are the number of data in training set
X_test = X_test.reshape(12630, 32, 32, 1)
X_val = X_val.reshape(4410, 32, 32, 1)
```

```python
from keras.preprocessing.image import ImageDataGenerator
 
datagen = ImageDataGenerator(width_shift_range=0.1,
                            height_shift_range=0.1,
                            zoom_range=0.2,
                            shear_range=0.1,
                            rotation_range=10.)
 
datagen.fit(X_train)
```

```python
# for X_batch, y_batch in
```

```python
batches = datagen.flow(X_train, y_train, batch_size = 15)#flow is the method to request our data generator to create more images, batch_size=# of images each time the generator is called
X_batch, y_batch = next(batches)#next func can mak a new set of 20 images on batches
 
fig, axs = plt.subplots(1, 15, figsize=(20, 5))
fig.tight_layout()
 
for i in range(15):
    axs[i].imshow(X_batch[i].reshape(32, 32))
    axs[i].axis("off")
 
print(X_batch.shape)
```

```python
y_train = to_categorical(y_train, 43)
y_test = to_categorical(y_test, 43)
y_val = to_categorical(y_val, 43)
```

```python
# create model
 
def modified_model():
  model = Sequential()
  model.add(Conv2D(60, (5, 5), input_shape=(32, 32, 1), activation='relu'))#60 are th number of filters, more filters demand more computer power
  #(5 ,5) tupls defines the dimention of our filter, each image is 32 32 size so (5 to 5) filter is a good size 60*5*5 +60(bias values) adjustabl valus
  #imp: (5,5)means that each 5 to 5 figure will become a 1 to 1 box all cells mutiplied to the weight values then summation divided to 25, so higher filter size smaller the final feature image
  model.add(Conv2D(60, (5, 5), activation='relu'))#to solve overfitting and validation (set) accuracy problem we can add more convolotional layers(another solution was to decrease learning rate)
  #by using extra convolotional layers the dimention of our data will diminish and less total parameters will remain to adjust
  model.add(MaxPooling2D(pool_size=(2, 2)))#it still has the depth of 30 (#of kernels filters) as pooling does not affect the depth
  
  model.add(Conv2D(30, (3, 3), activation='relu'))#in here 30*30*3*3+30
  model.add(Conv2D(30, (3, 3), activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  #by adding a dropout layer here we will block good neorons to let weak neorons train themselves better, eventually it will increase the validation accuracy 
  model.add(Flatten())#by flatten then our model is ready to fed into fully connected layers
  #very important: from here everything is like deep neural network, so conventional neural network will come before deep learning neural network which uses dense layers
  model.add(Dense(500, activation='relu'))
  model.add(Dropout(0.5))
  model.add(Dense(43, activation='softmax'))
  
  model.compile(Adam(lr = 0.001), loss='categorical_crossentropy', metrics=['accuracy'])#small learning rate of 0.001 has been used to increase the accuracy of model
  return model
```

```python
model = modified_model()
print(model.summary())
 #instead of model.fit we use model.fit_generator to make new versions of pictures out of our current data, doing rotations zoom in and out, ... so we can compensate shortage in data and make a better prediction 
 # adding this step is crucial when we don't have enough data  
history = model.fit_generator(datagen.flow(X_train, y_train, batch_size=50),
                            steps_per_epoch=2000,
                            epochs=10,
                            validation_data=(X_val, y_val), shuffle = 1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.xlabel('epoch')
 
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.legend(['training','test'])
plt.title('Accuracy')
plt.xlabel('epoch')
 
# TODO: Evaluate model on test data
score = model.evaluate(X_test, y_test, verbose=0)
```

```python
print('Test score:', score[0])
print('Test accuracy:', score[1])
 
#predict internet number
import requests
from PIL import Image
url = 'https://c8.alamy.com/comp/A0RX23/cars-and-automobiles-must-turn-left-ahead-sign-A0RX23.jpg'
r = requests.get(url, stream=True)
img = Image.open(r.raw)
plt.imshow(img, cmap=plt.get_cmap('gray'))
 
img = np.asarray(img)
img = cv2.resize(img, (32, 32))
img = preprocess(img)
plt.imshow(img, cmap = plt.get_cmap('gray'))
print(img.shape)
img = img.reshape(1, 32, 32, 1)
 
print("predicted sign: "+ str(model.predict_classes(img)))
```
