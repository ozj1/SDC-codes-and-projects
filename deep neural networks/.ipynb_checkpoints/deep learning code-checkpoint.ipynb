```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets# sklearn can give us a more detailed and complex datasets
import keras# again we use keras to predict our model
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
np.random.seed(0)
```

```python
n_pts = 500
X, y = datasets.make_circles(n_samples=n_pts, random_state = 123, noise=0.1, factor=0.2)#noise is the standard deviation, higher noise make it more difficult for our model to get the final classification tool
#diameter of inner section has 20 percent radius of of larger radius
print(X) 
print(y)#print lables
 
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='sigmoid'))#4 stands for 4 nodes in hidden layer section meaning 4 hidden layers to make a nonlinear (circle) boundary separation line, using more than 1 is bad as it make overfitting
model.add(Dense(1, activation='sigmoid'))
# 4 hidden layers after 2 layer of input then 1 layer for the output comes next 
model.compile(Adam(lr = 0.01), 'binary_crossentropy', metrics=['accuracy'])
```

```python
h=model.fit(x=X,y=y,verbose=1,batch_size=20,epochs=100,shuffle='true')
```

```python
#whn you put several plt.plot in one cell, notbook will put all in one graph
plt.plot(h.history['accuracy'])
plt.legend(['accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
 
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.title('loss')
plt.xlabel('epoch')
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 0.25, max(X[:,0]) + 0.25)
    y_span = np.linspace(min(X[:,1]) - 0.25, max(X[:,1]) + 0.25)
    xx, yy = np.meshgrid(x_span, y_span)
    grid = np.c_[xx.ravel(), yy.ravel()]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
 
x = 0.75
y = 0.75
 
point = np.array([[x, y]])
predict = model.predict(point)
plt.plot([x], [y], marker='o', markersize=10, color="red")
print("Prediction is: ", predict)


```
