```python
#whole point: overfitting is incorrect, the best fit is when it only gets general features of a picture not all the details, the details belong to that specific picture and can not be able to capture and predict new data results, we should focus on common feature among pictures not the details, if we focus on details then the fitted model is incorrect although it can work for that specific picture
#in training set, the purpose is to minimiz th error so overfitting happens, because of that validation set will be used to get the hyperparameters based on the general trend
 #so imp: how our code sees an image?: it will see each pixel in a scale of 0 to 255, each color in gray scale has a number and number is what machin learning program can see
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.datasets import mnist#import handwriting numbers
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical#as we are dealing with a multiclass datasets, and we need to do one hot encoding [0,0,1]
import random#random library will be used somewhere in the code
```

```python
np.random.seed(0)#seed=0 to get the same number whenever called
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data() #tuples for assigning output to different inputs, we get x data and y lable of training and test sets from mnist
 
print(X_train.shape)
print(X_test.shape)
print(y_train.shape[0])
```

```python
assert(X_train.shape[0] == y_train.shape[0]), "The number of images is not equal to the number of labels."
#assert is a tool for debugging, when the command is not corrct then the warning will be shown here
assert(X_test.shape[0] == y_test.shape[0]), "The number of images is not equal to the number of labels."
assert(X_train.shape[1:] == (28,28)), "The dimensions of the images are not 28x28"#1:=1 and 2 dimensions
assert(X_test.shape[1:] == (28,28)), "The dimensions of the images are not 28x28"
```

```python
num_of_samples = []
 
cols = 5
num_classes = 10
 
fig, axs = plt.subplots(nrows=num_classes, ncols = cols, figsize=(5, 8))#as it seems fig are frames and axs are th pics in them
fig.tight_layout()#prevent subplots from overlapping
for i in range(cols):#soton be soton mire jolo in cach soton j row ha ro mishmore va tasvireshon ro mishone to subplot marbote
    for j in range(num_classes):
        x_selected = X_train[y_train == j]#all the X_train data with a same lable number (for example all are 1 or 3)
        axs[j][i].imshow(x_selected[random.randint(0, len(x_selected - 1)), :, :], cmap=plt.get_cmap("gray"))
        axs[j][i].axis("off")#removing the axis
        if i == 2:#i is for colomns and j for rows
            axs[j][i].set_title(str(j))#string of j value
            num_of_samples.append(len(x_selected))
```

```python
#everything we have done so far is became familiar with mnist dataset,now we want to prepare our data to train our model
print(num_of_samples)
plt.figure(figsize=(12, 4))
plt.bar(range(0, num_classes), num_of_samples)
plt.title("Distribution of the training dataset")
plt.xlabel("Class number")
plt.ylabel("Number of images")
 
y_train = to_categorical(y_train, 10)#for one hot encoding we need ylables and number of lables or classes which is 10
y_test = to_categorical(y_test, 10)
 #one step to prepare our data is to normalize them, each image containes a pixel intensities from 0 upto 255, maximum pixel value of 255 will be normalized to th maximum value of 1
    #higher variance of 255 will make it difficult for machine learning algorithms to learn fast and quickly, that's why normalization is important
#in sigmoid and softmax function exp(1)=2.7 exp(255)=5.5E110, it mak it difficult for ML model to find bias, weight and hyperparameter values
X_train = X_train/255 
X_test = X_test/255
 # in order to use our data we must make them one dimentional to be able to multiplied to wight matrix 
num_pixels = 784
X_train = X_train.reshape(X_train.shape[0], num_pixels)#each image is 28pixels*28pixels=784pixels in total
X_test = X_test.reshape(X_test.shape[0], num_pixels)
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(10, input_dim=num_pixels, activation='relu'))#a hidden layer with 10 nodes
    model.add(Dense(30, activation='relu'))#another hidden layer with 30 nodes
    #if we increase the number of nodes it will make it easier for our model to fit the data and have smaller errors(so it's not good and will be a lot differnt than ) 
    model.add(Dense(10, activation='relu'))#another hidden layer with 10 nodes
    model.add(Dense(num_classes, activation='softmax'))#output layer 
    model.compile(Adam(lr=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
    return model
 
 #our input layer has 784 nodes if it was for larger images than 28 pixels then only conventional neural network would have been a computational way to solve and learn the data
model = create_model()
print(model.summary())
 
history = model.fit(X_train, y_train, validation_split=0.1, epochs = 10, batch_size = 200, verbose = 1, shuffle = 1)#validation set is 10% of the whole training set
 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.title('Loss')
plt.xlabel('epoch')
 
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['accuracy', 'val_accuracy'])
plt.title('Accuracy')
plt.xlabel('epoch')
 
score = model.evaluate(X_test, y_test, verbose=0)
print(type(score))
print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
import requests
from PIL import Image#pyton imaging library
 
url = 'https://www.researchgate.net/profile/Jose_Sempere/publication/221258631/figure/fig1/AS:305526891139075@1449854695342/Handwritten-digit-2.png'
response = requests.get(url, stream=True)
img = Image.open(response.raw)
plt.imshow(img, cmap=plt.get_cmap('gray'))
 
import cv2
 
img = np.asarray(img)
img = cv2.resize(img, (28, 28))#we use cv2 to resize the image
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#BGR2GRAY=blue green red to gray
img = cv2.bitwise_not(img)#the original image has background of white and number of black here it should be vise versa so bitwise will do that for us by converting the scale
plt.imshow(img, cmap=plt.get_cmap('gray'))# without cmap=plt.get_cmap('gray') the background is yellow
 #so imp: how our code sees an image?: it will see each pixel in a scale of 0 to 255, each collour in gray scale has a number and number is what machin learning program can see
img = img/255
img = img.reshape(1, 784)
 
prediction = model.predict_classes(img)
print("predicted digit:", str(prediction))
```
